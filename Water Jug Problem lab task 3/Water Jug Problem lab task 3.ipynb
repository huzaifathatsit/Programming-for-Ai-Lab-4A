```python
capacity_A = 4
capacity_B = 3
GOAL = 2
visited = set()

initial_state = (0, 0)
print(f"Step 1: Initial State: {initial_state}")
```

```python
def is_goal(state):
    return state[0] == GOAL

if is_goal(initial_state):
    print("Step 2: Goal Reached at Initial State!")
else:
    print("Step 2: Initial state is not the goal.")
```

```python
def get_next_states(state):
    x, y = state
    next_states = []
    # Rule 1: Fill Jug A
    next_states.append(((capacity_A, y), "Fill Jug A"))
    # Rule 2: Fill Jug B
    next_states.append(((x, capacity_B), "Fill Jug B"))
    # Rule 3: Empty Jug A
    next_states.append(((0, y), "Empty Jug A"))
    # Rule 4: Empty Jug B
    next_states.append(((x, 0), "Empty Jug B"))
    # Rule 5: Pour Jug A into Jug B
    transfer = min(x, capacity_B - y)
    next_states.append(((x - transfer, y + transfer), "Pour Jug A into Jug B"))
    # Rule 6: Pour Jug B into Jug A
    transfer = min(y, capacity_A - x)
    next_states.append(((x + transfer, y - transfer), "Pour Jug B into Jug A"))
    return next_states

for idx, (state, action) in enumerate(get_next_states(initial_state), start=1):
    print(f"Step 3.{idx}: {action} -> {state}")
```

```python
def dfs(state, path):
    if state in visited:
        return False
    visited.add(state)
    print(f"DFS Visiting: {state}")
    if is_goal(state):
        print("Goal Reached!")
        for step in path:
            print(step)
        return True
    for next_state, action in get_next_states(state):
        if next_state not in visited:
            print(f"Rule Applied: {action}")
            print(f"Next State: {next_state}")
            if dfs(next_state, path + [f"{action} -> {next_state}"]):
                return True
    return False

dfs(initial_state, [])
```
